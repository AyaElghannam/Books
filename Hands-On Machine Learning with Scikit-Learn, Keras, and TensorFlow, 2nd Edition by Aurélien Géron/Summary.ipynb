# Summary
## Definition
Machine Learning is the field of study that gives computers the ability to learn
without being explicitly programmed.

And a more engineering-oriented one:
A computer program is said to learn from experience E with respect to some task T
and some performance measure P, if its performance on T, as measured by P,
improves with experience E.

## Why ?

- Problems for which existing solutions require a lot of fine-tuning or long lists of rules: one Machine Learning algorithm can often simplify code and perform better than the traditional approach.
- Complex problems for which using a traditional approach yields no good solution: the best Machine Learning techniques can perhaps find a solution.
- Fluctuating environments: a Machine Learning system can adapt to new data.
- Getting insights about complex problems and large amounts of data

## Types

### Whether or not they are trained with human supervision
#### Supervised
In supervised learning, the training set you feed to the algorithm includes the desired solutions, called labels.
In other words we are given a data set and already know what our output correctly looks like i.e. there is a relation between i/p and o/p
##### Regression
We are trying to predict results within continous output, 'Mapping i/p variable to some continous function'
##### Classification
we are trying to predict a discrete results in a discrete output 'Mapping i/p variables into categories'
##### Algorithms
- k-Nearest Neighbors
- Linear Regression
- Logistic Regression
- Support Vector Machines (SVMs)
- Decision Trees and Random Forests
- Neural networks
Note that some regression algorithms can be used for classification as well, and vice versa. 
#### Unsupervised
In unsupervised learning, as you might guess, the training data is unlabeled
Allows us to approach problems with little or no idea what our results should look like , we can derive structure from data where we don't know the effect of variables , by clustering the data based on relationship among the variables i the data , NO feedback based on prediction results

ing data using a dimensionality reduction algorithm before you
feed it to another Machine Learning algorithm (such as a super
vised learning algorithm). It will run much faster, the data will take
up less disk and memory space, and in some cases it may also per
